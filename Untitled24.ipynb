```python
from typing import TypedDict

class AgentState(TypedDict):
    topic: str
    research: str
    output: str
```

```python
def researcher_agent(state):
    topic = state["topic"]

    research_text = f"""
    Research about {topic}:
    - Key concepts explained
    - Real-world applications
    - Benefits and challenges
    """

    print("🔍 Researcher: Research completed")

    state["research"] = research_text
    return state


def writer_agent(state):
    research = state["research"]

    article = f"""
    === ARTICLE ===

    Title: {state['topic']}

    {research}

    Conclusion:
    This topic is highly important in modern technology.
    """

    print("✍️ Writer: Article generated")

    state["output"] = article
    return state
```

```python
def supervisor(state):
    if state["research"] == "":
        print(" Supervisor → Researcher")
        return "researcher"

    elif len(state["research"]) < 120:
        print(" Supervisor → Researcher (loop)")
        return "researcher"

    else:
        print(" Supervisor → Writer")
        return "writer"
```

```python
from langgraph.graph import StateGraph, END

builder = StateGraph(AgentState)

# Add nodes
builder.add_node("researcher", researcher_agent)
builder.add_node("writer", writer_agent)

# Entry point
builder.set_entry_point("researcher")

# Conditional routing
builder.add_conditional_edges(
    "researcher",
    supervisor
)

# Final edge
builder.add_edge("writer", END)

graph = builder.compile()
```

```python
initial_state = {
    "topic": "Artificial Intelligence in Healthcare",
    "research": "",
    "output": ""
}

result = graph.invoke(initial_state)

print("\n FINAL OUTPUT:\n")
print(result["output"])
```
